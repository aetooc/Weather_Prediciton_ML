```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import precision_score, f1_score
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("data.csv")
df
```

```python
#Encoding the strings to Numericals
outlookCol=LabelEncoder()
tempCol=LabelEncoder()
humCol=LabelEncoder()
winCol=LabelEncoder()
```

```python
#Dropping the target variable and make it is as newframe
inputs=df.drop('Play',axis='columns')
target=df['Play']
target
```

```python
#Creating the new dataframe
inputs['outlook_n']= outlookCol.fit_transform(inputs['Outlook'])
inputs['Temp_n']= tempCol.fit_transform(inputs['Temp'])
inputs['Hum_n']= humCol.fit_transform(inputs['Humidity'])
inputs['win_n']= winCol.fit_transform(inputs['Windy'])
inputs
```

```python
#Dropping the string values
inputs_n=inputs.drop(['Outlook','Temp','Humidity','Windy'],axis='columns')
inputs_n
```

```python
#Applying the Gaussian naivebayes
classifier = GaussianNB()
classifier.fit(inputs_n,target)
```

```python
#85% accuracy 
classifier.score(inputs_n,target)
```

```python
# Use the classifier to make predictions on the entire dataset
y_pred = classifier.predict(inputs_n)

# Calculate the precision 
precision_yes = precision_score(target, y_pred, pos_label='yes')
precision_no = precision_score(target, y_pred, pos_label='no')
precision_avg = (precision_yes + precision_no) / 2


print("Precision for 'yes' class:", precision_yes)
print("Precision for 'no' class:", precision_no)
print("Average precision:", precision_avg)
```
